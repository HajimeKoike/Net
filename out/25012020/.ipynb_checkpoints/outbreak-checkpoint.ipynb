```python
epicurves=pd.DataFrame(columns=['Time'])
epicurves['Time']=np.arange(0,99,1)
for err in [5,10,20,50,100,200,500]:
    epi=pd.DataFrame(columns=[0,1])
    for time in range(99):
        df=pd.read_csv(f'./{dir}/{err}/{time}_pf.csv')
        count=pd.Series(df['State'].value_counts(),index=epi.columns)
        epi=epi.append(count,ignore_index=True)
    epicurves[f'I_{err/1000}']=epi[1]
```

```python
d=pd.read_csv(f'./{dir}/{100}/0_pf.csv')
d
```

```python
df=pd.read_csv(f'./{dir}/nature.csv',header=None)
epicurves['Nature']=df.sum(axis=1)
```

```python
epicurves
```

```python
epicurves.fillna(0,inplace=True)
```

```python
epicurves
```

```python
rmse_pf=pd.DataFrame(columns=['err','RMSE'])
for err in [5,10,20,50,100,200,500]:
    s=0
    for index,row in epicurves[f'I_{err/1000}'].iteritems():
        s+=(row-epicurves['Nature'].iloc[index])**2
    s=np.sqrt(s/epicurves.index.size)
    rmse=pd.Series([err*0.001,s],index=rmse_pf.columns)
    rmse_pf=rmse_pf.append(rmse,ignore_index=True)
```

```python
rmse_pf
```

```python
rmse_err=pd.DataFrame(columns=['err','RMSE'])
for err in [5,10,20,50,100,200,500]:
    s=0
    for index,row in epi_curve[f'epi_{err}'].iteritems():
        s+=(row-epi_curve['Nature'].iloc[index])**2
    s=np.sqrt(s/epi_curve.index.size)
    rmse=pd.Series([err*0.001,s],index=rmse_err.columns)
    rmse_err=rmse_err.append(rmse,ignore_index=True)
```

```python
rmse_err
```

```python
ax=rmse_err.plot.scatter(x='err',y='RMSE',c='blue',label='obs')
rmse_pf.plot.scatter(x='err',y='RMSE',c='red',ax=ax,label='pf')
plt.xscale('log')
plt.yscale('log')
plt.xlim([0.001,1])
plt.title('log(obs error) vs log(RMSE)')
plt.legend()
plt.savefig('logerr.png')
```

```python
for sample in range(1,100,10):
    epicurves=pd.DataFrame(columns=['Time'])
    epicurves['Time']=np.arange(0,99,1)
    size=500
    epi_curve=pd.DataFrame(columns=[0,1])
    for time in range(99):
        df=pd.read_csv(f'./{sample}/{size}_{time}_pf.csv')
        count=pd.Series(df['State'].value_counts(),index=epi_curve.columns)
        epi_curve=epi_curve.append(count,ignore_index=True)
    epicurves[f'I_{size}']=epi_curve[1]
```

```python
epicurves
```

```python
rmse_size=pd.DataFrame(columns=['Size','RMSE'])
for size in range(100,1100,100):
    s=0
    for index,row in epicurves[f'I_{size}'].iteritems():
        s+=(row-epicurves['Nature'].iloc[index])**2
    s=np.sqrt(s/epicurves.index.size)
    rmse=pd.Series([size,s],index=rmse_size.columns)
    rmse_size=rmse_size.append(rmse,ignore_index=True)
```
